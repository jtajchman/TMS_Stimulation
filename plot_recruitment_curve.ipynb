```python
from tms_probability.probability_analysis import get_thresholds, thresholds_to_probability, aggregate_threshold_probabilities
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
efields = np.linspace(180, 600, 50)

monophasic_fname = 'L5_LBC_cNAC_0'#'L5_TTPC2_cADpyr_0', 'L5_LBC_cNAC_0'
monophasic_results = get_thresholds(f'C:/Users/jacob/Documents/Work/NetPyNE/TMS_Stimulation/data/tms_thresholds/{monophasic_fname}/{monophasic_fname}_results.json')
monophasic_probabilities = [aggregate_threshold_probabilities(monophasic_results, efield) for efield in efields]

half_sine_fname = 'L5_LBC_cNAC_1'#'L5_TTPC2_cADpyr_1', 'L5_LBC_cNAC_1'
half_sine_results = get_thresholds(f'C:/Users/jacob/Documents/Work/NetPyNE/TMS_Stimulation/data/tms_thresholds/{half_sine_fname}/{half_sine_fname}_results.json')
half_sine_probabilities = [aggregate_threshold_probabilities(half_sine_results, efield) for efield in efields]

import matplotlib
matplotlib.rc('figure', figsize=(12, 6))
plt.plot(efields, monophasic_probabilities, efields, half_sine_probabilities)
plt.title('L5 Inh Cells', fontsize=20)
plt.xlabel('Electric Field Amplitude (V/m)', fontsize=15)
plt.ylabel('Recruitment Probability', fontsize=15)
plt.legend(['Monophasic (M0.2)', 'Half-Sine (M0.8)'], fontsize=15)
```
